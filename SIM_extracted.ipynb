```python
#v1.1
#Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python

# 1. Define Indian stock and market index
stock_symbols=  ['RELIANCE.NS', 'INFY.NS']    # Reliance Industries and Infosys
market_symbol = '^NSEI'         # NIFTY 50 index

# 2. Download historical data
start_date = '2025-01-01'
end_date = '2025-12-31'

stock_data = yf.download(stock_symbols, start=start_date, end=end_date)['Close']
market_data = yf.download(market_symbol, start=start_date, end=end_date)['Close']
print(stock_data.head())
print(market_data.head())
```

```python
# 3. Calculate daily returns
stock_returns = stock_data.pct_change().dropna()
market_returns = market_data.pct_change().dropna()

# Align dates
returns_df = stock_returns.join(market_returns).dropna()
#rename coumn to market
returns_df = returns_df.rename(columns={market_symbol: 'Market'})
print(returns_df.head())
```

```python
# 4. Run regression to estimate intercept and beta for RIL - This is market model and not SIM

X = sm.add_constant(returns_df['Market'])  # add intercept
y = returns_df[['RELIANCE.NS']] # Select both stock columns for y
model = sm.OLS(y, X, hasconst=True).fit()

alpha_ril = model.params['const']
beta_ril = model.params['Market']
r_squared = model.rsquared

print(model.summary())

print(f"Alpha: {alpha_ril:.4f}, Beta: {beta_ril:.4f}, R²: {r_squared:0.4f}")
```

```python
# 4. Run regression to estimate intercept and beta for Infy

X = sm.add_constant(returns_df['Market'])  # add intercept
y = returns_df[['INFY.NS']] # Select both stock columns for y
model = sm.OLS(y, X, hasconst=True).fit()

alpha_infy = model.params['const']
beta_infy = model.params['Market']
r_squared = model.rsquared

print(model.summary())

print(f"Alpha: {alpha_infy:.4f}, Beta: {beta_infy:.4f}, R²: {r_squared:0.4f}")
```
